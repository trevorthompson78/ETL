```python
split_data = billboard['Artist'].str.split("\xa0")
data = split_data.to_list() 
name = ['Artist','1','2','3','4','5']
new_df = pd.DataFrame(data, columns=name)
```

```python
spotify['Title'].unique()
```

```python
split_data = spotify['Title'].str.split("(")
data = split_data.to_list() 
data
name = ['Title','1']
new_df = pd.DataFrame(data, columns=name)
new_df = new_df.drop(['1'], axis=1)
new_df
```

```python
spotify_title= new_df.iloc[:,0]
spotify["Title"]=spotify_title
# spotify = spotify.drop(["Title"], axis=1)
spotify.reset_index()
```

```python
spotify.rename(columns={"Artist(s)":"Artist"}, inplace=True)
spotify.rename(columns={"No":"Chart_Number"}, inplace=True)
spotify.reset_index()
```

```python
billboard.rename(columns={"Artist(s)":"Artist", "No":"Chart_Number"}, inplace=True)
billboard
```

```python
connection_string = "postgres:postgres@localhost:5432/ETL"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
spotify.to_sql(name='spotify', con=engine, if_exists='append', index=True)
```

```python
billboard.to_sql(name='billboard', con=engine, if_exists='append', index=True)
```
