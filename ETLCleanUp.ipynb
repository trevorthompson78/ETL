```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
billboard = pd.read_csv('billboard_hot_100.csv')
billboard
```

```python
spotify = pd.read_csv('spotify_top_50.csv')
spotify
```

```python
spotify = spotify.drop(['Genre', 'Beats.Per.Minute', 'Energy', 'Danceability', 'Loudness..dB..', 'Liveness','Valence.', 'Length.','Acousticness..','Speechiness.','Popularity'],axis=1)
```

```python
spotify = spotify.rename(columns = {'Unnamed: 0':'No.','Track.Name':'Title', 'Artist.Name':'Artist(s)'})
```

```python
split_data = billboard['Artist'].str.split("\xa0")
data = split_data.to_list() 
name = ['Artist','1','2','3','4','5']
new_df = pd.DataFrame(data, columns=name)
```

```python
new_df.head()
new_df = new_df.drop(['1','2','3','4','5'], axis=1)
new_df.head()
```

```python
billboard_artist= new_df.iloc[:,0]
billboard["Artist(s)"]=billboard_artist
billboard = billboard.drop(["Artist"], axis=1)
```

```python
billboard["Title"]=billboard.iloc[:, 1].str.replace('"', '')
```
